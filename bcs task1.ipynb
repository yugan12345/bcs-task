```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#task 2. Read the data from the csv file using pandas data frame.
```

```python
df=pd.read_csv('Downloads/ipl.csv')
```

```python
#task 3. Check if there is any null value present in the data, if present handle them.
```

```python
df.isnull().sum()
```

```python
#8 null values in avg are null
```

```python
import math
median=math.floor(df.Avg.median())
df.Avg=df.Avg.fillna(median)                 
```

```python
#task5. Now you must perform linear regression on the number of 4s.
#task6. Create two data frames, one for features ( you may choose whichever columns you
#find suitable) and one for target (here target is number of 4s).
```

```python
features=df[['SR','Runs']]
target=df['4s']
```

```python
#task 7. Now split the dataset using train_test_split utility from sklearn into training and
#testing with the number of 4s as a target.
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target)
```

```python
#task 8. Now train the linear regression model using sklearn on train data and predict the
#values for test data.
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
prediction_test=model.predict(x_test)
prediction_test
prediction_train=model.predict(x_train)
```

```python
#task 9. Find mean square error on train data predictions and test data predictions
```

```python
msetest=mean_squared_error(y_test,prediction_test)
msetrain=mean_squared_error( y_train,prediction_train)
```
